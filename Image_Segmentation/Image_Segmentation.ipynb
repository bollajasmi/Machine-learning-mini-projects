```python
from skimage import io
from sklearn import cluster
import matplotlib.pyplot as plt
```

```python
image = io.imread('/content/image.lung.jpeg', as_gray=True)/255.0
h, w = image.shape
```

```python
image_2d = image.reshape(h*w,1)
```

```python
numcolors = 3
```

```python
kmeans_cluster = cluster.KMeans(n_clusters=int(numcolors), n_init='auto', random_state=0)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
newimage_float = cluster_centers[cluster_labels].reshape(h, w)
# Scale the float image to 0-255 range for better visualization
newimage = ((newimage_float - newimage_float.min()) / (newimage_float.max() - newimage_float.min()) * 255).astype('uint8')
```

```python
io.imshow(newimage)
io.show()
```

```python
plt.imshow(newimage)
plt.show()
```

```python
io.imsave('lung_kmeans.gif', newimage)
```
